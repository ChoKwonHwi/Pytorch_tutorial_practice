```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4 + 3*x**3 - 20*x**2
```

```python
x = np.linspace(-5,5,100)
plt.plot(x,f(x));
```

```python
results = optimize.minimize(f, -1)
results
```

```python
x_opt = results.x
```

```python
plt.plot(x,f(x));
plt.plot(x_opt,f(x_opt),'ro');
```

```python
def f(x):
    return x**2 - 20 * np.cos(x)

fig, axes = plt.subplots(2, 5, figsize=(18, 6))
idx = 0
xx = np.arange(-10, 10, 2)
x = np.linspace(-10, 10, 100)
for i, ax in enumerate(axes.flat) :
    point = xx[idx]
    results = optimize.minimize(f, point)
    x_opt = results.x
    ax.plot(x,f(x))
    ax.plot(x_opt,f(x_opt),'ro')
    idx +=1
```
